```python
# Hamza Jadoon
# Task 1 - Iris Data Exploration
# This project explores the Iris dataset to understand patterns,
# feature relationships, and data distribution using basic visualization.

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")

from google.colab import files
uploaded = files.upload()

df = pd.read_csv("Iris.csv")

print("Dataset Loaded\n")

# basic info
print("Shape:")
print(df.shape)
print("\n")

print("Columns:")
print(df.columns)
print("\n")

print("First rows:")
print(df.head())
print("\n")

# dataset details
print("Info:")
print(df.info())
print("\n")

print("Summary:")
print(df.describe())
print("\n")

# drop id
if "Id" in df.columns:
    df.drop("Id", axis=1, inplace=True)

# missing check
print("Missing values:")
print(df.isnull().sum())
print("\n")

# scatter plot
plt.figure(figsize=(8,6))
sns.scatterplot(
    x="SepalLengthCm",
    y="SepalWidthCm",
    hue="Species",
    data=df
)
plt.title("Sepal Length vs Sepal Width")
plt.show()

# pair plots
sns.pairplot(df, hue="Species")
plt.show()

# histograms
df.hist(figsize=(10,8))
plt.suptitle("Feature Distributions", fontsize=16)
plt.show()

# boxplots
plt.figure(figsize=(12,6))

for i, column in enumerate(df.columns[:-1]):
    plt.subplot(2,2,i+1)
    sns.boxplot(y=df[column])
    plt.title(column)

plt.tight_layout()
plt.show()

# correlation
plt.figure(figsize=(8,6))
sns.heatmap(
    df.corr(numeric_only=True),
    annot=True,
    cmap="coolwarm"
)
plt.title("Correlation Matrix")
plt.show()

print("Observations:")
print("""
- 150 samples, 4 features
- Petal features highly related
- Species mostly separable
- Very few outliers
- Data clean and ready
""")
```
